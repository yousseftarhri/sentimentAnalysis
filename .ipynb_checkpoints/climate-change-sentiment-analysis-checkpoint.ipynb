# Load Data

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/twiteer-sentiment/data.csv")
df.tail()
```

# Transform the Data

```python
#This line processes a column named review_rating where each entry contains text like "4.5 out of 5 stars". The goal is to extract the numeric rating and convert it into an integer.
df["review_rating"] = df["review_rating"].apply(lambda x: int(float(x.split()[0])))
#This line processes a column named review_comment, ensuring that no comment exceeds 512 characters. If a comment is longer than 512 characters, it will be truncated.
df["review_comment"] = df["review_comment"].apply(lambda x: x[:512] if len(x) > 512 else x)
```

```python
#Labels: 0 -> Negative; 1 -> Neutral; 2 -> Positive
def rating_sentiment(rating):
    if rating > 3:
        return 2
    elif rating < 3:
        return 0
    else:
        return 1

df["review_rating"] = df["review_rating"].apply(rating_sentiment)
```

**Clean Data**

```python
import re
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Remove unwanted characters (if necessary)
    # text = re.sub(r'[^a-zA-Z\s]', '', text)  # Uncomment this line if you want to remove non-alphabet characters

    return text


# Clean the review_comment column
df['review_comment'] = df['review_comment'].apply(clean_text)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Bar chart for sentiment analysis
sentiment_counts = df["review_rating"].value_counts()

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)

# Set the title and labels
plt.title('Sentiment Analysis Results')
plt.xlabel('Sentiment')
plt.ylabel('Count')

# Set custom labels for the x-axis
labels = [ 'Positive (2)', 'Neutral (1)','Negative (0)']
plt.xticks(ticks=sentiment_counts.index, labels=labels)

# Show the plot
plt.show()
```

# Topic Modeling

**Preprocess the Text Data**

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim import corpora

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Preprocess the reviews
stop_words = set(stopwords.words('english'))

def preprocess(text):
    # Tokenize
    tokens = word_tokenize(text.lower())
    # Remove stop words and non-alphabetic tokens
    return [word for word in tokens if word.isalpha() and word not in stop_words]

df['processed_reviews'] = df['review_comment'].apply(preprocess)
```

**Create a Dictionary and Corpus**

```python
# Create a dictionary representation of the documents
dictionary = corpora.Dictionary(df['processed_reviews'])

# Create a corpus
corpus = [dictionary.doc2bow(review) for review in df['processed_reviews']]
```

**Train the LDA Model**

```python
from gensim.models import LdaModel

# Set parameters
num_topics = 3  # Adjust the number of topics as needed
passes = 15

# Train LDA model
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=passes)
```

**Identify Topics**

```python
# Print the topics
for idx, topic in lda_model.print_topics(num_words=5):
    print(f"Topic {idx + 1}: {topic}")
```

# Named Entity Recognition (NER)

```python
import spacy
from spacy import displacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")

# Sample text data
reviews = [
    "I bought the new iPhone 14 from Apple Store in New York.",
    "The experience at Amazon was fantastic. Their delivery is always on time!",
    "I recommend using the Google Pixel for photography.",
    "I visited the Eiffel Tower in Paris last summer.",
    "Microsoft is launching a new product next month."
]

reviews = list(df["review_comment"])

# Process the reviews and extract entities
for review in reviews:
    doc = nlp(review)
    print(f"Review: {review}")
    for ent in doc.ents:
        print(f" - {ent.text} ({ent.label_})")

# Visualize entities for the first review
doc = nlp(reviews[0])  # Process the first review
displacy.render(doc, style='ent', jupyter=True)
```
